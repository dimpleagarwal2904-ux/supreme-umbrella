```python
#create an interactive user input game where user guesses a random number

import random

def guessing_game():
    randomnumber= random.randint(1,100) #returns a random integer number 1<= N <= 100
    guesses = 0
    guessed= False
    print(" I am thinking of a number between 1 and 100. Can you guess it?")
    while(guessed== False):
        guess= int(input("Enter your guess between 1 and 100: "))
        guesses+=1
        if (guess >= 1 and guess <=100):
            if(guess == randomnumber):
                print(f"Correct guess in {guesses} tries")
                guessed= True
            elif(guess < randomnumber):
                print("Too low!")
            else:
                print("Too high!")
        else:
            print("Value error. Number not between 1 and 100")
    
    



print("=====Number Guessing Game=====")
while True:
    guessing_game()
    play_again= input("Do you want to try again? (yes/no): ")
    if(play_again != 'yes'):
        print("Thank you for playing!")
        break
```
