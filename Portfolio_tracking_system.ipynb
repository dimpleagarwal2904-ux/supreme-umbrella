```python
#Write a program that calculates the value of each stock 
# holding in a portfolio, computes the total portfolio 
# value, lists all tickers, and finds the stock with the 
# highest price.
```

```python
portfolio={
    "holdings":[
        {"ticker":"AAPL", "shares": 10, "price": 150.50},
        {"ticker":"GOOGL","shares":5,"price":2800.00},
        {"ticker":"MSFT","shares":8,"price":310.00}
    ],
    "total_value":0
}

total=0
for holding in portfolio["holdings"]:
    holding_value= holding["shares"]*holding["price"]
    total+= holding_value
    print(f"{holding['ticker']} having {holding['shares']} shares @${holding['price']} = ${holding_value:.2f}")
    #.2f tells python to format number as floating type value with exactly 2 digits after decimal
    #inside an f-string it is written as (value: format (.2f in this case))

portfolio["total_value"] = total
print(f" \n Total portfolio value is $ {portfolio['total_value']:.2f}")

tickers= [h["ticker"] for h in portfolio["holdings"]]
print(f"Holdings: {tickers}")

highest= max(portfolio["holdings"], key=lambda x: x["price"])
print(f"Highest priced: {highest['ticker']} at $ {highest['price']} ")
```
