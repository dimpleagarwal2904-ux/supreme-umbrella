```python
#Build a command-line calculator CLI that performs 
# arithmetic operations

def add(a,b):
    return a+b

def subtract(a,b):
    if(a>b):
        return a-b
    else:
        return b-a

def multiply(a,b):
    return a*b

def divide(a,b):
    if(b != 0):
        return a/b
    else:
        return "Error: can't divide by 0"  

def calculator():
    print("=====Simple Calculator=====")
    while True:
        print("\n Select operation")
        print("1. Add \n 2. Subtract \n 3. Multipy \n 4. Divide \n 5. Exit")

        choice= input("Enter your choice: ")

        if(choice=='5'):
            print("Thank you for using Calculator!")
            break
        elif( choice in ('1','2','3','4')):   #input takes string commands, hences 1,2,3,4 in quotes
            #in command checks membership in lists, tuples, sets, strings, dictionaries
            num1= float(input("Enter your first nuumber: "))
            num2= float(input("Enter your second number: "))
            if(choice=='1'):
               print("The sum is:", add(num1, num2))
            elif(choice == '2'):
                print("The difference is: ", subtract(num1, num2))
            elif(choice=='3'):
                print("The product is: ", multiply(num1, num2))
            else:
                print("The quotient is: ", divide(num1, num2))
        else:
            print("Invalid input")

calculator()
```
