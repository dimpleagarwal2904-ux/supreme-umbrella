```python
# Prints out 0,1,2,3,4 and then it prints "count value reached 5"
count=0
while (count<5):
    print(count)
    count+=1
    if(count ==10):
        break
else:
    print("Count value reached 5")
#break can be only inside while loop. after while, else 
#gets executed anyway unless there is a break statement inside while
#that gets executed.

```

```python
#1. Add a function named list_benefits() that returns the following list of 
# strings: "More organized code", "More readable code", "Easier code reuse", 
# "Allowing programmers to share and connect code together"

#2. Add a function named build_sentence(info) which receives a single argument
# containing a string and returns a sentence starting with the given string and 
# ending with the string " is a benefit of functions!"

def list_benefits():
    return["More organized code", "More readable code", "Easier code reuse", 
           "Allowing programmers to share and connect code together"]

def build_sentence(arg):
    return "%s is a benefit of functions!"%(arg)

def name_the_benefits_of_functions():
    list_of_benefits = list_benefits()
    for arg in list_of_benefits:
        print(build_sentence(arg))

name_the_benefits_of_functions()        
```
